# Pandas Case Study - Kashti Dataset

```python
# Importing Libraries

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns 
```

```python
# Importing Datset

kashti = pd.read_csv("Dataset\Titanic.csv")
kashti.head(20)
```

```python
# Basic Statistics or Summary

kashti.describe()
```

```python
# Dropping few columns and creating new dataset

new_kashti = kashti.drop(["PassengerId","Ticket"], axis=1)
new_kashti.head()
```

```python
# Calculating Mean
new_kashti.mean()
```

```python
# Counting no. of survived people
new_kashti.value_counts(["Survived"])
```

```python
new_kashti.groupby(["Sex"]).mean()
```

```python
new_kashti.groupby(["Sex","Pclass"]).mean()
```

```python
# Children
new_kashti[new_kashti["Age"] < 18].groupby(["Sex","Pclass"]).mean()
```
